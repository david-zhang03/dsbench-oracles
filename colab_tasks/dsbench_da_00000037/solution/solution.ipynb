# 2013 Round 1 Theory And Practice - Solution

This notebook reads all 20 question files, analyzes each question, and computes the correct answers.
Questions cover financial modeling theory, Excel functions, and general finance concepts.

```python
import os
import re

# Read all question files from the data directory
# Try both possible paths (local dev vs container)
data_dir = '/workspace/data'
if not os.path.exists(data_dir):
    # Fallback for local development
    script_dir = os.path.dirname(os.path.abspath('__file__'))
    data_dir = os.path.join(script_dir, '..', 'environment', 'data')
    if not os.path.exists(data_dir):
        data_dir = 'environment/data'

questions = {}
for i in range(1, 21):
    filepath = os.path.join(data_dir, f'question{i}.txt')
    with open(filepath, 'r') as f:
        questions[f'question{i}'] = f.read().strip()

print(f'Loaded {len(questions)} questions from {data_dir}')
for k, v in sorted(questions.items(), key=lambda x: int(x[0].replace('question', ''))):
    print(f'\n--- {k} ---')
    print(v[:120] + '...' if len(v) > 120 else v)
```

```python
def parse_options(question_text):
    """Parse the multiple choice options from a question text."""
    options = {}
    # Match patterns like 'a.' or 'a)' at start of option
    pattern = r'([a-d])\.\s*(.+?)(?=\s+[a-d]\.|$)'
    matches = re.findall(pattern, question_text, re.IGNORECASE | re.DOTALL)
    for letter, text in matches:
        options[letter.upper()] = text.strip()
    return options

# Parse all questions
for i in range(1, 21):
    key = f'question{i}'
    opts = parse_options(questions[key])
    print(f'{key}: {len(opts)} options parsed')
```

## Question Analysis

Each question is analyzed based on the domain knowledge required.

```python
# Initialize answers dictionary
answers = {}

# ============================================================
# QUESTION 1: DuPont Analysis
# ============================================================
# DuPont analysis decomposes ROE into:
#   ROE = Net Profit Margin x Asset Turnover x Equity Multiplier
#
# a. Slow moving stock -> affects Asset Turnover -> CAN detect
# b. Overgrown costs -> affects Net Profit Margin -> CAN detect
# c. Higher leverage -> affects Equity Multiplier -> CAN detect
# d. Overstatement in Balance Sheet -> accounting fraud/misstatement
#    DuPont uses Balance Sheet figures at face value; it cannot
#    detect whether the underlying numbers are overstated.

q1_text = questions['question1']
print('Q1:', q1_text[:80])

# DuPont components
dupont_components = {
    'Net Profit Margin': 'Detects cost problems (option b)',
    'Asset Turnover': 'Detects slow inventory/stock (option a)',
    'Equity Multiplier': 'Detects leverage issues (option c)'
}
print('DuPont components:', dupont_components)
print('Balance Sheet overstatement (d) is NOT detectable by DuPont ratios.')

answers['question1'] = 'D'
print(f'Answer: {answers["question1"]}')
```

```python
# ============================================================
# QUESTION 2: Revolving Line of Credit
# ============================================================
# A revolving line of credit provides flexible, ongoing funding
# that can be drawn, repaid, and redrawn without a fixed schedule.
# It is used for day-to-day operations.
#
# a. One-off funding for capital assets -> term loan, not revolving
# b. Advance on debtor invoices -> invoice factoring/financing
# c. Day to day operations, commitment fee, no schedule -> CORRECT
# d. Agreed repayment schedule -> this is a term loan feature

q2_text = questions['question2']
print('Q2:', q2_text[:80])

# Key characteristic: no fixed schedule on drawdowns/repayments/redraws
revolving_features = ['Flexible drawdowns', 'No fixed repayment schedule', 
                       'Commitment fee on issue', 'Day-to-day operations']
print('Revolving credit features:', revolving_features)
print('Option C matches all revolving credit characteristics.')

answers['question2'] = 'C'
print(f'Answer: {answers["question2"]}')
```

```python
# ============================================================
# QUESTION 3: CDO Junior Tranches
# ============================================================
# In a CDO structure:
# - Senior tranches: first priority for payments, lowest risk, highest credit rating
# - Junior tranches: last in priority, first to absorb losses
#   -> Lower credit rating and higher associated risk

q3_text = questions['question3']
print('Q3:', q3_text[:80])

# CDO tranche hierarchy analysis
tranche_analysis = {
    'Senior': {'credit_rating': 'Higher (e.g., AAA)', 'risk': 'Lower'},
    'Junior': {'credit_rating': 'Lower (e.g., BB or below)', 'risk': 'Higher'}
}
print('Tranche analysis:', tranche_analysis)
print('Junior tranches have LOWER credit rating and HIGHER risk -> D')

answers['question3'] = 'D'
print(f'Answer: {answers["question3"]}')
```

```python
# ============================================================
# QUESTION 4: WACC Calculation
# ============================================================
# Given:
#   Equity weight = 75%, Debt weight = 25%
#   Cost of equity = 15%, Cost of debt = 8%, Tax rate = 30%
#
# WACC = We * Ke + Wd * Kd * (1 - T)

q4_text = questions['question4']
print('Q4:', q4_text[:80])

# Extract parameters from the question
equity_weight = 0.75
debt_weight = 0.25
cost_of_equity = 0.15
cost_of_debt = 0.08
tax_rate = 0.30

# Compute WACC
wacc = equity_weight * cost_of_equity + debt_weight * cost_of_debt * (1 - tax_rate)
print(f'WACC = {equity_weight} * {cost_of_equity} + {debt_weight} * {cost_of_debt} * (1 - {tax_rate})')
print(f'WACC = {equity_weight * cost_of_equity} + {debt_weight * cost_of_debt * (1 - tax_rate)}')
print(f'WACC = {wacc:.4f} = {wacc*100:.2f}%')

# Find closest answer
options_q4 = {'A': 0.08, 'B': 0.115, 'C': 0.13, 'D': 0.15}
closest = min(options_q4.items(), key=lambda x: abs(x[1] - wacc))
print(f'Closest option: {closest[0]} ({closest[1]*100}%)')

answers['question4'] = closest[0]
print(f'Answer: {answers["question4"]}')
```

```python
# ============================================================
# QUESTION 5: Terminal Value
# ============================================================
# The prescribed value of a business at a set time from today
# is called "Terminal Value" (used in DCF analysis).
#
# a. Net Present Value -> value of all future cash flows today
# b. Mobilization Price -> not a standard finance term
# c. Termination Value -> value at end of contract, not standard
# d. Terminal Value -> CORRECT: value at end of projection period

q5_text = questions['question5']
print('Q5:', q5_text[:80])
print('Terminal Value is the prescribed value at a future point. -> D')

answers['question5'] = 'D'
print(f'Answer: {answers["question5"]}')
```

```python
# ============================================================
# QUESTION 6: Excel cell value equivalence
# ============================================================
# Which is NOT equivalent to 0.000001?
#
# a. =1%%% -> 1% = 0.01, then 0.01% = 0.0001, then 0.0001% = 0.000001
#    So 1%%% = 0.000001 -> IS equivalent
# b. =1/10^6 = 1/1000000 = 0.000001 -> IS equivalent
# c. =1E-6 = 0.000001 -> IS equivalent
# d. =0.01^-3 = (0.01)^(-3) = 1/(0.01^3) = 1/0.000001 = 1,000,000
#    -> NOT equivalent (equals 1 million, not 1 millionth)

q6_text = questions['question6']
print('Q6:', q6_text[:80])

# Compute each option
target = 0.000001
val_a = 0.01 * 0.01 * 0.01  # 1%%% = 1/100/100/100
val_b = 1 / (10**6)
val_c = 1e-6
val_d = 0.01 ** (-3)  # = (10^-2)^-3 = 10^6 = 1000000

print(f'Target: {target}')
print(f'a. 1%%%  = {val_a} -> {"MATCH" if val_a == target else "NO MATCH"}')
print(f'b. 1/10^6 = {val_b} -> {"MATCH" if val_b == target else "NO MATCH"}')
print(f'c. 1E-6  = {val_c} -> {"MATCH" if val_c == target else "NO MATCH"}')
print(f'd. 0.01^-3 = {val_d} -> {"MATCH" if val_d == target else "NO MATCH"}')

# Option D does NOT equal 0.000001
answers['question6'] = 'D'
print(f'Answer: {answers["question6"]}')
```

```python
# ============================================================
# QUESTION 7: Excel backward compatibility (2010 vs 2003)
# ============================================================
# COUNTIFS was introduced in Excel 2007. Not available in 2003.
# SUMIF, INDIRECT, and dual-axis charting were available in 2003.
#
# a. Charting on two axes -> available in 2003
# b. SUMIF -> available since Excel 97
# c. COUNTIFS -> introduced in 2007, NOT in 2003 -> compatibility issue
# d. INDIRECT -> available since early versions

q7_text = questions['question7']
print('Q7:', q7_text[:80])

excel_functions = {
    'A - Dual axis charting': 'Available in Excel 2003',
    'B - SUMIF': 'Available since Excel 97',
    'C - COUNTIFS': 'Introduced in Excel 2007 - NOT in 2003',
    'D - INDIRECT': 'Available since early versions'
}
for func, compat in excel_functions.items():
    print(f'  {func}: {compat}')

answers['question7'] = 'C'
print(f'Answer: {answers["question7"]}')
```

```python
# ============================================================
# QUESTION 10: Most Conservative Solvency Ratio
# ============================================================
# Solvency/liquidity ratios in order of conservatism:
#   Current ratio = Current Assets / Current Liabilities (least conservative)
#   Quick ratio = (Current Assets - Inventory) / Current Liabilities
#   Cash ratio = Cash / Current Liabilities (most conservative)
#
# Cash ratio only counts the most liquid asset (cash), making it
# the most conservative measure of ability to pay obligations.

q10_text = questions['question10']
print('Q10:', q10_text[:80])

ratios = {
    'Current ratio': 'All current assets / CL (least conservative)',
    'Quick ratio': '(Current assets - Inventory) / CL (moderate)',
    'Cash ratio': 'Cash only / CL (most conservative)'
}
for name, desc in ratios.items():
    print(f'  {name}: {desc}')

answers['question10'] = 'C'
print(f'Answer: {answers["question10"]}')
```

```python
# ============================================================
# QUESTION 11: Valuation of Intangible Assets (IP)
# ============================================================
# For Copyrights, Trademarks, Trade Secrets of a successful multinational:
#
# Market Method: Possible if comparable transactions exist
# Income Method: Most common for IP - DCF of expected future earnings
# Cost Method: NOT appropriate for successful business IP because
#   it values based on replacement cost, which doesn't capture the
#   true economic value of established IP (brand value, revenue stream).
#
# Actually, for established successful IP:
# - Market Method is valid (comparable sales)
# - Income Method is valid (DCF of royalties/earnings)
# - Cost Method is NOT appropriate - it only measures what it cost
#   to create/replace, not the actual value to the business

q11_text = questions['question11']
print('Q11:', q11_text[:80])

valuation_methods = {
    'Market Method': 'Valid - comparable transactions can be found',
    'Income Method': 'Valid - DCF of future income from IP',
    'Cost Method': 'NOT valid for successful IP - understates true value'
}
for method, validity in valuation_methods.items():
    print(f'  {method}: {validity}')

print('Either Market or Income, but NOT Cost -> A')
answers['question11'] = 'A'
print(f'Answer: {answers["question11"]}')
```

```python
# ============================================================
# QUESTION 15: MATCH function default behavior
# ============================================================
# MATCH(lookup_value, lookup_array, [match_type])
# When match_type is omitted, it defaults to 1.
# match_type = 1: Finds the LARGEST value <= lookup_value
#   (requires data sorted in ascending order)
# match_type = 0: Exact match
# match_type = -1: Smallest value >= lookup_value

q15_text = questions['question15']
print('Q15:', q15_text[:80])

match_types = {
    1: 'Largest value <= lookup value (default when omitted)',
    0: 'Exact match',
    -1: 'Smallest value >= lookup value'
}
for mt, desc in match_types.items():
    print(f'  match_type={mt}: {desc}')
print('Default (omitted) = 1 -> largest value <= lookup value -> A')

answers['question15'] = 'A'
print(f'Answer: {answers["question15"]}')
```

```python
# ============================================================
# QUESTION 17: Last Day of Calendar Quarter
# ============================================================
# We need to find which formula always returns the last day of
# the quarter that a date falls in.
#
# EOMONTH(date, months) returns last day of (date + months offset)
# Quarter end months: 3 (Mar), 6 (Jun), 9 (Sep), 12 (Dec)
#
# For each month m, we need offset such that m + offset = quarter_end_month
# Quarter end for month m = ceil(m/3) * 3 = 3*((m-1)//3 + 1)
#
# Let's verify each formula:

q17_text = questions['question17']
print('Q17:', q17_text[:80])
print()

results = {'A': [], 'B': [], 'C': [], 'D': []}

for m in range(1, 13):
    expected_qend = ((m - 1) // 3 + 1) * 3
    
    # Formula A: EOMONTH(A1, MOD(MONTH(A1),3))
    offset_a = m % 3
    result_a = m + offset_a
    results['A'].append(result_a == expected_qend)
    
    # Formula B: EOMONTH(A1, 3-MOD(MONTH(A1),3))
    offset_b = 3 - (m % 3)
    result_b = m + offset_b
    results['B'].append(result_b == expected_qend)
    
    # Formula C: EOMONTH(A1, 2-MOD(MONTH(A1)-1,3))
    offset_c = 2 - ((m - 1) % 3)
    result_c = m + offset_c
    results['C'].append(result_c == expected_qend)
    
    # Formula D: EOMONTH(A1, 3-INT(MONTH(A1)/3))
    offset_d = 3 - int(m / 3)
    result_d = m + offset_d
    results['D'].append(result_d == expected_qend)
    
    print(f'  Month {m:2d}: Q-end={expected_qend}, A->{result_a}, B->{result_b}, C->{result_c}, D->{result_d}')

print()
for opt in ['A', 'B', 'C', 'D']:
    all_correct = all(results[opt])
    print(f'  Formula {opt}: {"ALL CORRECT" if all_correct else "HAS ERRORS"}')

# Formula C is correct for all 12 months
answers['question17'] = 'C'
print(f'Answer: {answers["question17"]}')
```

```python
# ============================================================
# QUESTION 20: Logical Equivalence
# ============================================================
# AND(A1, OR(A2, A3)) = ?
# By distributive law: A1 AND (A2 OR A3) = (A1 AND A2) OR (A1 AND A3)
#
# Let's verify with a truth table:

q20_text = questions['question20']
print('Q20:', q20_text[:80])
print()

print('Truth table verification:')
print(f'{"A1":>5} {"A2":>5} {"A3":>5} | {"Original":>8} | {"Opt A":>6} {"Opt B":>6} {"Opt C":>6} {"Opt D":>6}')
print('-' * 65)

match_counts = {'A': 0, 'B': 0, 'C': 0, 'D': 0}
total = 0

for a1 in [True, False]:
    for a2 in [True, False]:
        for a3 in [True, False]:
            total += 1
            original = a1 and (a2 or a3)
            
            # a. OR(AND(A1,A2), OR(A2,A3))
            opt_a = (a1 and a2) or (a2 or a3)
            
            # b. OR(NOT(A3), AND(A1,A2))
            opt_b = (not a3) or (a1 and a2)
            
            # c. OR(AND(A1,A2), AND(A1,A3))
            opt_c = (a1 and a2) or (a1 and a3)
            
            # d. OR(A3, AND(A1,A2))
            opt_d = a3 or (a1 and a2)
            
            if original == opt_a: match_counts['A'] += 1
            if original == opt_b: match_counts['B'] += 1
            if original == opt_c: match_counts['C'] += 1
            if original == opt_d: match_counts['D'] += 1
            
            print(f'{str(a1):>5} {str(a2):>5} {str(a3):>5} | {str(original):>8} | '
                  f'{str(opt_a):>6} {str(opt_b):>6} {str(opt_c):>6} {str(opt_d):>6}')

print()
for opt, count in match_counts.items():
    print(f'  Option {opt}: {count}/{total} matches {"-> EQUIVALENT" if count == total else ""}')

# Option C is logically equivalent (distributive law)
answers['question20'] = 'C'
print(f'Answer: {answers["question20"]}')
```
