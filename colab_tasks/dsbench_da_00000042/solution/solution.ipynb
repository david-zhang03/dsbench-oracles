Identify the Excel function described in each of the 50 question files.

```python
import os
import re

# Data directory - the question files are here
DATA_DIR = os.path.join(os.path.dirname(os.getcwd()), 'environment', 'data')
if not os.path.isdir(DATA_DIR):
    # Try workspace path (for container environment)
    DATA_DIR = '/workspace/data'
if not os.path.isdir(DATA_DIR):
    # Fallback to relative from notebook
    DATA_DIR = '../environment/data'

print(f'Data directory: {DATA_DIR}')
print(f'Files found: {len([f for f in os.listdir(DATA_DIR) if f.startswith("question")])}')
```

```python
# Read the introduction to understand the task
intro_path = os.path.join(DATA_DIR, 'introduction.txt')
with open(intro_path, 'r') as f:
    intro = f.read().strip()
print('Introduction:', intro)
```

```python
# Read all 50 question files
questions = {}
for i in range(1, 51):
    qfile = os.path.join(DATA_DIR, f'question{i}.txt')
    with open(qfile, 'r') as f:
        raw = f.read().strip()
    # Strip the leading question number (e.g., "1 Joins several...")
    desc = re.sub(r'^\d+\s+', '', raw)
    questions[f'question{i}'] = desc

# Display the questions
for qkey in sorted(questions.keys(), key=lambda x: int(x.replace('question', ''))):
    print(f'{qkey}: {questions[qkey]}')
```

```python
def score_function(question_text, keywords):
    """Score how well a function's keywords match the question text.
    
    Longer keyword phrase matches are weighted more heavily to favor
    specific matches over generic ones.
    """
    question_lower = question_text.lower()
    score = 0
    for kw in keywords:
        if kw.lower() in question_lower:
            # Weight by keyword length - longer, more specific matches score higher
            score += len(kw)
    return score


def identify_function(question_text):
    """Identify which Excel function a question description matches.

    Uses keyword scoring against all known Excel functions and returns
    the function with the highest match score. Applies disambiguation
    rules for closely related functions (e.g., IRR vs MIRR vs XIRR).
    """
    best_func = None
    best_score = 0

    for func_name, keywords in EXCEL_FUNCTIONS.items():
        s = score_function(question_text, keywords)
        if s > best_score:
            best_score = s
            best_func = func_name

    # Disambiguation: if "modified" appears, prefer MIRR over IRR
    q_lower = question_text.lower()
    if best_func == "IRR" and "modified" in q_lower:
        best_func = "MIRR"

    return best_func


print('Matching functions defined.')
```

```python
# ────────────────────────────────────────────────────────────────────────────
# Apply the matching algorithm to all 50 questions
# ────────────────────────────────────────────────────────────────────────────

answers = {}

for i in range(1, 51):
    qkey = f'question{i}'
    desc = questions[qkey]
    func = identify_function(desc)
    answers[qkey] = func
    print(f'{qkey}: "{desc[:60]}..." -> {func}')

print(f'\nTotal answers: {len(answers)}')
print('\nFinal answers dict:')
for k in sorted(answers.keys(), key=lambda x: int(x.replace('question', ''))):
    print(f'  {k}: {answers[k]}')
```
