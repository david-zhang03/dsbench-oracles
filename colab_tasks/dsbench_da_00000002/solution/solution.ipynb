```python
import os
import re
import numpy as np

# Determine the data directory
# Support both /workspace/data (container) and local paths
DATA_DIR = None
for candidate in ['/workspace/data', os.path.join(os.path.dirname(os.path.abspath('.')), 'environment', 'data'),
                  os.path.join(os.getcwd(), '..', 'environment', 'data'),
                  '/home/ddzhang/DSBench/colab_tasks/dsbench_da_00000002/environment/data']:
    if os.path.isdir(candidate):
        DATA_DIR = candidate
        break

if DATA_DIR is None:
    raise FileNotFoundError('Could not find the data directory')

print(f'Using data directory: {DATA_DIR}')
```

```python
# Read all question files
questions = {}
for i in range(1, 26):
    filepath = os.path.join(DATA_DIR, f'question{i}.txt')
    with open(filepath, 'r') as f:
        questions[f'question{i}'] = f.read().strip()
    print(f'question{i}: {questions[f"question{i}"][:80]}...')

print(f'\nLoaded {len(questions)} questions.')
```

```python
# Parse question text into question body and options
def parse_question(text):
    """Parse a question text into the question body and option dict."""
    # Normalize the text
    text = text.strip()
    
    # Find options - they start with a. b. c. d. (case insensitive)
    # Split on option markers
    options = {}
    
    # Find all option positions
    # Pattern: space or start followed by a letter and period
    option_pattern = r'(?:^|\s)([a-dA-D])\.\s'
    matches = list(re.finditer(option_pattern, text))
    
    question_body = text[:matches[0].start()].strip() if matches else text
    
    for idx, match in enumerate(matches):
        letter = match.group(1).upper()
        start = match.end()
        end = matches[idx + 1].start() if idx + 1 < len(matches) else len(text)
        option_text = text[start:end].strip()
        options[letter] = option_text
    
    return question_body, options

# Test parsing
for i in [1, 14, 15]:
    body, opts = parse_question(questions[f'question{i}'])
    print(f'Q{i}: {body[:60]}...')
    for k, v in opts.items():
        print(f'  {k}: {v[:60]}')
    print()
```

```python
# Question 1: Financial Modeler's main objective when gathering INTERNAL information
# for a valuation.
# Internal info gathering -> thorough analysis of the company's operations/earnings.
# Option A is about examining financial statements (this is part of it but more specific)
# Option C is "Thorough analysis to gain an adequate understanding of the subject 
# company's operational management and earnings ability" - this is the main OBJECTIVE
# Answer: C

q1_text = questions['question1'].lower()
# The question asks about the "main objective" of gathering internal information.
# The key phrase is "thorough analysis" and "operational management and earnings ability"
body1, opts1 = parse_question(questions['question1'])

# Identify that the answer mentioning "thorough analysis" and "operational management" 
# is the main objective (not just examining documents or facilities)
answer1 = None
for letter, text in opts1.items():
    if 'thorough analysis' in text.lower() and 'operational management' in text.lower():
        answer1 = letter
        break
print(f'Q1 answer: {answer1}')
```

```python
# Question 14: Tom's Mining Equipment - which method is most appropriate?
# Earnings: 75400, 65200, 87600, 90500, 53900
# The earnings are volatile (no clear trend up or down), so unweighted average
# is most appropriate. Geometric progression needs consistent growth,
# trend line needs a clear trend.
# Answer: B (Unweighted average method)

# Actually compute the earnings pattern to determine the method
earnings = [75400, 65200, 87600, 90500, 53900]
earnings_arr = np.array(earnings, dtype=float)

# Check for consistent growth (geometric progression)
growth_rates = []
for i in range(1, len(earnings)):
    growth_rates.append((earnings[i] - earnings[i-1]) / earnings[i-1])
print(f'Growth rates: {growth_rates}')

# Check if growth rates are consistent (for geometric progression)
growth_std = np.std(growth_rates)
growth_mean = np.mean(growth_rates)
print(f'Growth rate mean: {growth_mean:.4f}, std: {growth_std:.4f}')

# Check for trend (correlation with time)
years = np.array([1, 2, 3, 4, 5], dtype=float)
correlation = np.corrcoef(years, earnings_arr)[0, 1]
print(f'Correlation with time: {correlation:.4f}')

# Earnings fluctuate significantly with no clear trend -> unweighted average is best
# Growth rates alternate positive/negative, correlation is weak
body14, opts14 = parse_question(questions['question14'])
answer14 = None
for letter, text in opts14.items():
    if 'unweighted average' in text.lower():
        answer14 = letter
        break
print(f'Q14 answer: {answer14}')
```
